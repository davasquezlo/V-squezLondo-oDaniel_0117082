```python
#Ejercicio 17
#%%

'''
 17. Leer una lista y determinar cuál es el segundo elemento más grande de 
     dicha lista. No utilice ningún tipo de algoritmo de ordenamiento para 
     realizar este ejercicio.
'''

import sys                                         #Se Importa la librería sys

lista=[]                                       #Se inicia con  la lista vacía
len1 = int(input('Digite el tamaño de la lista: '))
          #Se ingresa la cantidad de elementos de la lista, es decir su tamaño

while len1 < 0:
    len1 = int(input('ERROR-No dígito un valor válido para tamaño de lista'))
                            #Imprime "ERROR" si el valor ingresado es negativo
for i in range (len1):
    lista.append(input(f'Digite el elemento {i+1}: '))
                                         #Se ingresa cada elemento de la lista

print('La lista es: ',lista)                   #Se imprime la lista ingresada

tamaño_1 = 0                                    #Se inicia la variable en cero
for i in lista:
    tamaño = sys.getsizeof(i)      #Se determina el tamaño de la lista con sys
    if tamaño >= tamaño_1:     
        tamaño_1 = tamaño

for i in lista:
    if sys.getsizeof(i) == tamaño_1:
        lista.remove(i)               #Se elimina el elemento con mayor tamaño
        
tamaño_2 = 0
for i in lista:
    tamaño = sys.getsizeof(i) #Se vuelve a iterar la lista y volver a buscar
                             #el elemento con el tamaño más grande, que para el 
                          #ejercicio en general es el segundo tamaño más grande 
    if tamaño > tamaño_2:      
        tamaño_2 = tamaño      
        
for i in lista:
    if sys.getsizeof(i)==tamaño_1:

        print('El segundo elemento con mayor tamaño es', i)
                     #Imprime el segundo elemento de la lista con mayor tamaño
                     
```

```python
#Ejercicio 11
#%%

'''
 11. Escribir un programa que reciba una cadena de texto y reporte:
a) cuántas letras vocales en mayúscula se entraron;
b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
c) cuántos dígitos se entraron,
d) cuántos espacios se entraron;
e) cuántas palabras reservadas se entraron.
    Utilizar para ello los métodos de cadenas especializados como
    .isupper(), .isdigit(), etc. y la palabra reservada in.
'''

import keyword                                         #Se importa la librería 

texto = input('Digite una cadena de texto: ')  #Se ingresa una cadena de texto

vocales = 'AEIOUÜ'
tildes = 'áéíóúÁÉÍÓÚ'  

num_vocales_may = 0                             #Se inicia la variable en cero
num_tildes = 0                                  #Se inicia la variable en cero
num_digitos = 0                                 #Se inicia la variable en cero
num_espacios = 0                                #Se inicia la variable en cero
num_reservadas = 0                              #Se inicia la variable en cero

for i in texto:
    if i in vocales:
        num_vocales_may = num_vocales_may + 1
        
    elif i in tildes:
        num_tildes = num_tildes + 1
        
    elif i.isdigit():
        num_digitos = num_digitos + 1
        
    elif i == ' ':
        num_espacios = num_espacios + 1       
     #Se determina el número de vocales, letras con tildes, dígitos o espacios

texto_list = texto.split(' ') #Se divide la cadena en palabras con los espacios

for i in range(len(texto_list)):
    
    if texto_list[i] in keyword.kwlist:         #Palabras reservadas de Python
        num_reservadas = num_reservadas + 1
    
print(f'En la cadena de texto ingresada hay: ')
print(f'{num_vocales_may} vocales en mayúscula')
print(f'{num_tildes} letras con tilde')
print(f'{num_digitos} dígitos')
print(f'{num_espacios} espacios')
print(f'{num_reservadas} palabras reservadas')
                                                   #Se imprimen los resultados
```

```python
#Ejercicio 18
#%%

'''
 18. Leer una matriz cuadrada A, obtener su matriz triangular superior asociada
     U y determinar qué fila de U contiene la mayor cantidad de unos.
'''

n=int(input('Digite el orden de la matriz cuadrada: '))
                             #Se digita el orden de la matriz cuadrada (n = m)

while n<=0:
    n=int(input('ERROR-No dígito un valor válido para orden de la matriz'))   
                            #Imprime "ERROR" si el valor ingresado es negativo
print('Matriz cuadrada de orden', n)   #Imprime el orden de la matriz cuadrada

matriz=[]                                                     #Se inicia vacía
j=0                                                                  #Columnas
while j<n:
    fila=[]
    i=0                                                                 #filas
    while i<n:
        while True:
            try:             #Se ingresan los valores para completar la matriz
                num=float(input(f'Ingrese el número {j+1,i+1}=  '))
                break
            except ValueError:
                print('NO DEBE COLOCAR LETRAS') #Imprime no poder escribir 
                                       #letras o cadenas de texto en la matriz
        i=i+1
        fila.append(num)
    matriz.append(fila)
 
    j= j+1
    
print('\nLa matriz que usted registró es la siguiente:')

#Se imprime la matriz resultante
for fila in matriz:
    print(fila)

U=matriz                         #Se determina la matriz triangular asociada U
for i in range (n):
    for j in range(n):
        if j>=i:
            (U[i])[j]=(matriz[i])[j] 
        else:
            (U[i])[j]=0
            
print('\nLa siguiente es la matriz triangular superior U')           
for fila in U:
    print(fila)

print('\nSe presenta la lista de cantidad de unos presentes en cada \
fila de forma ordenada')      #Se analiza fila por fila(lista por lista) de la 
                              #matriz para determinar cuantos unos posee cada 
                              #una y mostrar la que más unos tenga
unos=[]
for i in range (n):
     conta= U[i].count(1)
     unos.append(conta)
print(unos)

mayor=unos[0]
for g in range(len(unos)):
    if unos[g]>mayor:
        mayor=unos[g]
pos=[]
for g in range(len(unos)):
    if unos[g] == mayor:
        pos.append((g+1))
        
print('\nLa mayor cantidad de unos es', mayor,' y se encuentra\
en la(s) fila(s)',pos)  #Se imprime la mayor cantidad de unos y una lista con 
                        #el número de fila o filas que contiene tal cantidad
```

```python
#Ejercicio 19
#%%

'''
 19. Leer una lista e identifique la sublista más larga dentro de dicha lista
     que se encuentra ordenada ascendentemente. Por ejemplo, en la lista 
     [5, 6, 5, 1, 2, 6, 9, 2, 4, 1] la mayor sublista ordenada es la 
     [1, 2, 6, 9].
'''

lista = []                                      #Se inicia con  la lista vacía

tamaño = int(input('Digite la cantidad de elementos de la lista (tamaño): '))
          #Se ingresa la cantidad de elementos de la lista, es decir su tamaño
if tamaño < 0:
    print('ERROR-No dígito un valor válido para tamaño de lista')
                            #Imprime "ERROR" si el valor ingresado es negativo
while tamaño > 0:
        lista.append((int(input('Digite la lista de números: '))))
        tamaño = tamaño - 1
                                         #Se ingresa cada elemento de la lista
listas = []                                     #Se inicia con  la lista vacía

for i in range(len(lista)):
    aux = i       
    aux_2 = -1  #Variables auxiliares para la identación de listas ascendentes
                
    for j in range(i, len(lista)-1): #Separa las sublistas de forma ascendente
        if lista[j] <= lista[j+1]:
            aux_2 = j+1
        else:
            break
    listas.append(lista[aux:aux_2+1])
    
lista_mayor = listas[0]
for i in listas:                   
    if len(i) >= len(lista_mayor):
        lista_mayor = i   #Determina y separa la sublista ascendente mas larga
    
print('La lista ingresada es: ', lista)

if len(lista_mayor) == 0:
    print('No se halló una lista.')   #Imprime si no hay sublistas ascendentes
else:
    print(f'La sublista más larga es: {lista_mayor}')
                                      #Imprime la sublista ascendente esperada
```

```python
#Ejercicio 20
#%%

'''
 20. Hacer un programa que lea una matriz y la rote 90 grados en sentido 
     antihorario.
'''
 
import numpy as np                               #Se importa la librería numpy

while True:              #Se ingresa el tamaño de la matriz (filas y columnas)
    try:
        m = int(input('Ingrese la cantidad de filas: '))
        n = int(input('Ingrese la cantidad de columnas: '))
        break
    except ValueError:
        print('ERROR. Filas y columnas son cantidades enteras')

matriz = []            #Se ingresan los valores de la matriz termino a termino
j = 0                   
while j < m:           
    fil = []             
    i = 0               
    while i < n:
        while True:
            try:
                num = float(input(f'Ingrese un número {j+1,i+1}: '))
                break
            except ValueError:
                print('Recuerde que solo debe ingresar números')
                        #Imprime "ERROR" si el valor ingresado no es un número
        i += 1
        fil.append(num)
    matriz.append(fil)
    j += 1

print('\nLa matriz ingresada fue:') #Se imprime la matriz original digitada 
                                    #que se va a rotar
for fil in matriz:
    print(fil)
    
matriz = np.array(matriz)                            #Se convierte en un array

matriz_90 = []                                       #Se crea una matriz vacía 
for i in range(n-1,-1,-1):
    matriz_90.append(list(matriz[:,i]))
   
matriz_90 = np.array(matriz_90)

print('\nLa matriz rotada 90 grados es:')
print(matriz_90)         #Se imprime la matriz ingresada pero girada 90 grados
```
