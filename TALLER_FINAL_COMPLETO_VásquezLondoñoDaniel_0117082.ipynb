```python
#Ejercicio 2  
#%%

'''
 2. Leer un número entero y escribirlo en reversa. Por ejemplo, 461975 será
    579164. No utilizar comandos de ordenamiento.
'''

from math import floor #Importa la función floor de la librería math

x = input('Digite un número entero: ') #Asigna valor a la variable "x"
y = abs(int(x))              #Valor absoluto de tipo entero del valor anterior
rever = 0                    #Se empieza la variable "rever" desde cero

while y > 0:                       #Mientras la variable "y" sea mayor a cero
    residuo = y - (floor(y/10))*10 #Determina el último digito del valor
                                   #ingresado inicialmente
    rever = (rever * 10) + residuo #Lleva la cuenta para obtener el reverso
    y = int(y/10)                  #Reduce el valor ingresado inicialmente
                                   #para un nuevo ciclo
if int(x) >= 0:
    print('El número', x, 'escrito en reversa es: ',rever) #Imprime el número
                                                   #en reversa de un positivo
        
else:
    print('El número', x, 'escrito en reversa es: ',-rever) #Imprime el número
                                                    #en reversa de un negativo
```

```python
#Ejercicio 9
#%%

'''
 9. Hacer un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) que 
    corresponden al centro y al radio de dos círculos. Lea un punto de 
    coordenadas (a,b). Determina si (a,b) está contenido: a) dentro del 
    círculo 1; b) dentro del círculo 2; c) dentro de ambos círculos; d) fuera
    de ambos círculos. Recordemos que el círculo es el lugar geométrico de los 
    puntos del plano cuya distancia a otro punto fijo, llamado centro, es 
    menor o igual que una cantidad constante, llamada radio.
'''

coord_x1 = abs(int(input('Digite la coordenada x del circulo 1: ')))
coord_y1 = abs(int(input('Digite la coordenada y del circulo 1: ')))
radio_c1 = abs(int(input('Digite el radio del circulo 1: ')))
  #Se ingresan las coordenadas del centro del círculo 1 y longitud de su radio

coord_x2 = abs(int(input('Digite la coordenada x del circulo 2: ')))
coord_y2 = abs(int(input('Digite la coordenada y del circulo 2: ')))
radio_c2 = abs(int(input('Digite el radio del circulo 2: ')))
  #Se ingresan las coordenadas del centro del círculo 2 y longitud de su radio

punto_x = abs(int(input('Digite la coordenada x del punto: ')))
punto_y = abs(int(input('Digite la coordenada y del punto: ')))
                                        #Se ingresan las coordenadas del punto

dist_c1_punto = ((punto_x - coord_x1)**2 + (punto_y - coord_y1)**2)**(1/2)
dist_c2_punto = ((punto_x - coord_x2)**2 + (punto_y - coord_y2)**2)**(1/2)
        #Se calculan las distancias entre el centro de cada círculo y el punto

if (dist_c1_punto <= radio_c1) and (dist_c2_punto <= radio_c2): 
    print('El punto está contenido dentro de ambos círculos')
                         #Imprime si el punto está adentro de los dos círculos
    
elif (dist_c1_punto > radio_c1) and (dist_c2_punto > radio_c2):
    print('El punto está contenido fuera de ambos círculos')
                          #Imprime si el punto está afuera de los dos círculos
    
else:
    if (dist_c1_punto <= radio_c1):
        print('El punto está contenido dentro del círculo 1')
                          #Imprime si el punto solo está adentro del círculo 1
        
    if (dist_c2_punto <= radio_c2): 
        print('El punto está contenido dentro del círculo 2')
                          #Imprime si el punto solo está adentro del círculo 2
```

```python
#Ejercicio 10
#%%

'''
 10. Sin utilizar el método .upper() para cadenas, escribir un programa que 
     lea una cadena de texto y la imprima en mayúscula. Tenga en cuenta que el 
     programa debe aceptar las letras con tilde, la ñ y la ü.
'''

lista = list(input('Dígite una cadena de texto: '))
                                               #Se ingresa una cadena de texto
tamaño_lista = len(lista)                #Se determina la longitud de la lista
tamaño_lista_2 = tamaño_lista #Se hace una especie de copia del valor anterior
lista2 = []                                          #Se inicia la lista vacía
                                    
while tamaño_lista > 0: 
    
    termino = lista[tamaño_lista_2-tamaño_lista]   #Se identifica cada termino
    ubicación = ord(termino)                  #Determina el código del termino
    posicion_nueva = ubicación-32                 #Se vuelve Mayúscula (ASCII)
    term_nuevo = chr(posicion_nueva)        #Determina el caracter del termino
    lista2.append(term_nuevo)                       #Se adiciona a la la lista
    tamaño_lista = tamaño_lista - 1           #Se reduce el tamaño de la lista
    
cadena = ''.join(lista2)                       #Se convierte la lista a cadena

print('La cadena de texto en Mayúsculas es: ',cadena)
                                  #Se imprime la cadena de texto en Mayúsculas
```

```python
#Ejercicio 12
#%%

'''
 12. Hacer un programa que lea dos palabras (podrían estar en mayúscula o 
     minúscula) y determine cual está primero en el diccionario. El programa 
     debe soportar letras con tilde, la ü y la ñ.
'''

texto1 = input('Ingrese una palabra en Mayúscula o minúscula: ') 
                                     #Se ingresa una palabra o cadena de texto
texto2 = input('Ingrese una segunda palabra en Mayúscula o minúscula: ')
                                     #Se ingresa una palabra o cadena de texto
cadena_texto1 = texto1.lower()      #Se convierte a minúsculas la cadena texto
cadena_texto2 = texto2.lower()      #Se convierte a minúsculas la cadena texto

x = 'áéíóúü'
y = 'aeiouu'                          #Se crea una lista para no operar tildes

cambio = str.maketrans(x,y)           #Se hace el cambio para no operar tildes
cambio_t1 = cadena_texto1.translate(cambio)    #Se hace el cambio en las listas
cambio_t2 = cadena_texto2.translate(cambio)    #Se hace el cambio en las listas

if cambio_t1 < cambio_t2: 
    print('El texto',cambio_t1,'va primero en el diccionario que',\
          cambio_t2)                      #Se imprime el orden alfabeticamente
else:
    if cambio_t2 < cambio_t1:
        print('El texto',cambio_t2,'va primero en el diccionario que',\
              cambio_t1)                  #Se imprime el orden alfabeticamente
    else:                  
        print('Los dos textos son iguales por lo tanto tienen el mismo\
 en el diccionario')    #Se imprime el mismo lugar de ser iguales las palabras
```

```python
#Ejercicio 13
#%%

'''
 13. Escribir un programa que lea una cadena de texto y la imprima como un 
     triángulo. Por ejemplo, si la cadena entrada es: "Es una locura odiar a 
     todas las rosas porque una te pinchó: renunciar a todos tus sueños porque 
     uno de ellos no se realizó - El Principito", el programa debe imprimir:
'''

Triangulo = list(input("Ingrese un texto: ")) 
                                     #Se ingresa una palabra o cadena de texto
x = ''.join(Triangulo)                              #Se une la cadena de texto
y = x.replace(" ","")                        #Se eliminan todos los parentesis

Triangulo_lista = list(y)                                    #Se crea la lista
tamaño = len(Triangulo_lista)              #Se determina el tamaño de la lista

inicio = 0                            
final = 1
final_fin = 1                         #Se determinan las condiciones iniciales

elemento_1 = Triangulo_lista[inicio:final]
elemento_2 = ''.join(elemento_1)
elemento_3 = elemento_2.center(tamaño)     #Se determina el centro el elemento

print(elemento_3)

while tamaño >= inicio:   #Se crea un ciclo para las demas filas del Triángulo
    inicio = final_fin      #Siguiendo la progresión que describe el Triángulo
    final = final + 1                               #Se aumenta cada que itera
    final_fin = final**2
    elemento = Triangulo_lista[inicio:final_fin]
    elemento_1 =''.join(elemento)
    elemento_2 = elemento_1.center(tamaño)              #Y Centrando cada fila
    print(elemento_2)                                 #Se imprime el Triángulo
```

```python
#Ejercicio 18
#%%

'''
 18. Leer una matriz cuadrada A, obtener su matriz triangular superior asociada
     U y determinar qué fila de U contiene la mayor cantidad de unos.
'''

orden = int(input('Digite el orden de la matriz cuadrada: '))
                   #Se digita el orden de la matriz cuadrada (filas = columnas)
while orden <= 0:
    orden=int(input('ERROR-No dígito un valor válido para orden de la matriz'))   
                            #Imprime "ERROR" si el valor ingresado es negativo
print('Matriz cuadrada de orden', orden)#Imprime el orden de la matriz cuadrada

matriz = []                                                   #Se inicia vacía
j = 0                                                                #Columnas
while j < orden:
    fila = []
    i = 0                                                               #filas
    while i < orden:
        while True:
            try:             #Se ingresan los valores para completar la matriz
                num=float(input(f'Ingrese el número {j+1,i+1}=  '))
                break
            except ValueError:
                print('NO DEBE COLOCAR LETRAS') #Imprime no poder escribir 
                                       #letras o cadenas de texto en la matriz
        i = i + 1
        fila.append(num)
    matriz.append(fila)
 
    j = j + 1
    
print('\nLa matriz que usted registró es la siguiente:')

#Se imprime la matriz resultante
for fila in matriz:
    print(fila)

U = matriz                       #Se determina la matriz triangular asociada U
for i in range (orden):
    for j in range(orden):
        if j >= i:
            (U[i])[j] = (matriz[i])[j] 
        else:
            (U[i])[j] = 0
            
print('\nLa siguiente es la matriz triangular superior U')           
for fila in U:
    print(fila)

print('\nSe presenta la lista de cantidad de unos presentes en cada \
fila de forma ordenada')      #Se analiza fila por fila(lista por lista) de la 
                              #matriz para determinar cuantos unos posee cada 
                              #una y mostrar la que más unos tenga
unos = []
for i in range (orden):
     conta= U[i].count(1)
     unos.append(conta)
print(unos)

mayor = unos[0]
for g in range(len(unos)):
    if unos[g] > mayor:
        mayor = unos[g]
pos = []
for g in range(len(unos)):
    if unos[g] == mayor:
        pos.append((g + 1))
        
print('\nLa mayor cantidad de unos es', mayor,' y se encuentra\
 en la(s) fila(s)',pos)  #Se imprime la mayor cantidad de unos y una lista con 
                        #el número de fila o filas que contiene tal cantidad
```

```python
#Ejercicio 19
#%%

'''
 19. Leer una lista e identifique la sublista más larga dentro de dicha lista
     que se encuentra ordenada ascendentemente. Por ejemplo, en la lista 
     [5, 6, 5, 1, 2, 6, 9, 2, 4, 1] la mayor sublista ordenada es la 
     [1, 2, 6, 9].
'''

lista = []                                      #Se inicia con  la lista vacía

tamaño = int(input('Digite la cantidad de elementos de la lista (tamaño): '))
          #Se ingresa la cantidad de elementos de la lista, es decir su tamaño
if tamaño < 0:
    print('ERROR-No dígito un valor válido para tamaño de lista')
                            #Imprime "ERROR" si el valor ingresado es negativo
while tamaño > 0:
        lista.append((int(input('Digite la lista de números: '))))
        tamaño = tamaño - 1
                                         #Se ingresa cada elemento de la lista
listas = []                                     #Se inicia con  la lista vacía

for i in range(len(lista)):
    aux = i       
    aux_2 = -1  #Variables auxiliares para la identación de listas ascendentes
                
    for j in range(i, len(lista)-1): #Separa las sublistas de forma ascendente
        if lista[j] <= lista[j+1]:
            aux_2 = j+1
        else:
            break
    listas.append(lista[aux:aux_2+1])
    
lista_mayor = listas[0]
for i in listas:                   
    if len(i) >= len(lista_mayor):
        lista_mayor = i   #Determina y separa la sublista ascendente mas larga
    
print('La lista ingresada es: ', lista)

if len(lista_mayor) == 0:
    print('No se halló una lista.')   #Imprime si no hay sublistas ascendentes
else:
    print(f'La sublista más larga es: {lista_mayor}')
                                      #Imprime la sublista ascendente esperada
```

```python
#Ejercicio 20
#%%

'''
 20. Hacer un programa que lea una matriz y la rote 90 grados en sentido 
     antihorario.
'''
 
import numpy as np                               #Se importa la librería numpy

while True:              #Se ingresa el tamaño de la matriz (filas y columnas)
    try:
        filas = int(input('Ingrese la cantidad de filas: '))
        columnas = int(input('Ingrese la cantidad de columnas: '))
        break
    except ValueError:
        print('ERROR. Filas y columnas son cantidades enteras')

matriz = []            #Se ingresan los valores de la matriz termino a termino
x = 0                   
while x < filas:           
    fil = []                                         #Se inicia la lista vacía
    y = 0                                       #Se inicia la variable en cero       
    while y < columnas:
        while True:
            try:
                num = float(input(f'Ingrese un número {x+1,y+1}: '))
                break
            except ValueError:
                print('Recuerde que solo debe ingresar números')
                        #Imprime "ERROR" si el valor ingresado no es un número
        y = y + 1
        fil.append(num)
    matriz.append(fil)
    x = x + 1

print('\nLa matriz ingresada fue:') #Se imprime la matriz original digitada 
                                    #que se va a rotar
for fil in matriz:
    print(fil)
    
matriz = np.array(matriz)                            #Se convierte en un array

matriz_90 = []                                       #Se crea una matriz vacía 
for i in range(columnas-1,-1,-1):
    matriz_90.append(list(matriz[:,i]))
   
matriz_90 = np.array(matriz_90)

print('\nLa matriz rotada 90 grados es:')
print(matriz_90)         #Se imprime la matriz ingresada pero girada 90 grados
```
