```python
#Ejercicio 2  
#%%

'''
 2. Leer un número entero y escribirlo en reversa. Por ejemplo, 461975 será
    579164. No utilizar comandos de ordenamiento.
'''

from math import floor #Importa la función floor de la librería math

x = input('Digite un número entero: ') #Asigna valor a la variable "x"
y = abs(int(x))              #Valor absoluto de tipo entero del valor anterior
rever = 0                    #Se empieza la variable "rever" desde cero

while y > 0:                       #Mientras la variable "y" sea mayor a cero
    residuo = y - (floor(y/10))*10 #Determina el último digito del valor
                                   #ingresado inicialmente
    rever = (rever * 10) + residuo #Lleva la cuenta para obtener el reverso
    y = int(y/10)                  #Reduce el valor ingresado inicialmente
                                   #para un nuevo ciclo
if int(x) >= 0:
    print('El número', x, 'escrito en reversa es: ',rever) #Imprime el número
                                                   #en reversa de un positivo
        
else:
    print('El número', x, 'escrito en reversa es: ',-rever) #Imprime el número
                                                    #en reversa de un negativo
    
```

```python
#Ejercicio 9
#%%

'''
 9. Hacer un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) que 
    corresponden al centro y al radio de dos círculos. Lea un punto de 
    coordenadas (a,b). Determina si (a,b) está contenido: a) dentro del 
    círculo 1; b) dentro del círculo 2; c) dentro de ambos círculos; d) fuera
    de ambos círculos. Recordemos que el círculo es el lugar geométrico de los 
    puntos del plano cuya distancia a otro punto fijo, llamado centro, es 
    menor o igual que una cantidad constante, llamada radio.
'''

coord_x1 = abs(int(input('Digite la coordenada x del circulo 1: ')))
coord_y1 = abs(int(input('Digite la coordenada y del circulo 1: ')))
radio_c1 = abs(int(input('Digite el radio del circulo 1: ')))
  #Se ingresan las coordenadas del centro del círculo 1 y longitud de su radio

coord_x2 = abs(int(input('Digite la coordenada x del circulo 2: ')))
coord_y2 = abs(int(input('Digite la coordenada y del circulo 2: ')))
radio_c2 = abs(int(input('Digite el radio del circulo 2: ')))
  #Se ingresan las coordenadas del centro del círculo 2 y longitud de su radio

punto_x = abs(int(input('Digite la coordenada x del punto: ')))
punto_y = abs(int(input('Digite la coordenada y del punto: ')))
                                        #Se ingresan las coordenadas del punto

dist_c1_punto = ((punto_x - coord_x1)**2 + (punto_y - coord_y1)**2)**(1/2)
dist_c2_punto = ((punto_x - coord_x2)**2 + (punto_y - coord_y2)**2)**(1/2)
        #Se calculan las distancias entre el centro de cada círculo y el punto

if (dist_c1_punto <= radio_c1) and (dist_c2_punto <= radio_c2): 
    print('El punto está contenido dentro de ambos círculos')
                         #Imprime si el punto está adentro de los dos círculos
    
elif (dist_c1_punto > radio_c1) and (dist_c2_punto > radio_c2):
    print('El punto está contenido fuera de ambos círculos')
                          #Imprime si el punto está afuera de los dos círculos
    
else:
    if (dist_c1_punto <= radio_c1):
        print('El punto está contenido dentro del círculo 1')
                          #Imprime si el punto solo está adentro del círculo 1
        
    if (dist_c2_punto <= radio_c2): 
        print('El punto está contenido dentro del círculo 2')
                          #Imprime si el punto solo está adentro del círculo 2
```

```python
#Ejercicio 10
#%%

'''
 10. Sin utilizar el método .upper() para cadenas, escribir un programa que 
     lea una cadena de texto y la imprima en mayúscula. Tenga en cuenta que el 
     programa debe aceptar las letras con tilde, la ñ y la ü.
'''

lista = list(input('Dígite una cadena de texto: '))
                                               #Se ingresa una cadena de texto
tamaño_lista = len(lista)                #Se determina la longitud de la lista
tamaño_lista_2 = tamaño_lista #Se hace una especie de copia del valor anterior
lista2 = []                                          #Se inicia la lista vacía
                                    
while tamaño_lista > 0: 
    
    termino = lista[tamaño_lista_2-tamaño_lista]   #Se identifica cada termino
    ubicación = ord(termino)                  #Determina el código del termino
    posicion_nueva = ubicación-32                 #Se vuelve Mayúscula (ASCII)
    term_nuevo = chr(posicion_nueva)        #Determina el caracter del termino
    lista2.append(term_nuevo)                       #Se adiciona a la la lista
    tamaño_lista = tamaño_lista - 1           #Se reduce el tamaño de la lista
    
cadena = ''.join(lista2)                       #Se convierte la lista a cadena

print('La cadena de texto en Mayúsculas es: ',cadena)
                                  #Se imprime la cadena de texto en Mayúsculas
```
